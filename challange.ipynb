```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

### Criação e Manipulação do DataFrame

```python
# Dicionário de faturamento
dict_faturamento = {
    "data_ref": [
        "2023-01-01",
        "2020-02-01",
        "2021-03-01",
        "2022-04-01",
        "2023-05-01",
        "2023-06-01",
        "2020-07-01",
        "2021-08-01",
        "2022-09-01",
        "2023-10-01",
        "2022-11-01",
        "2023-12-01",
    ],
    "valor": [
        400000,
        890000,
        760000,
        430000,
        920000,
        340000,
        800000,
        500000,
        200000,
        900000,
        570000,
        995000,
    ],
}
```

```python
# Criação do DataFrame baseado nos dados do dicionário
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento.info()
```

```python
# Alterar o tipo da coluna data
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

```python
df_faturamento.info()
```

```python
df_faturamento['mes'] = df_faturamento['data_ref'].dt.strftime('%B')
```

```python
df_faturamento
```

### Calculo de medidas

```python
# Média de vendas
df_faturamento['valor'].mean()
```

```python
# Mediana
df_faturamento['valor'].median()
```

### Gráficos analíticos

```python
# Gráfico de barras vertical
df_faturamento.plot.bar(x='mes', y='valor', title="Faturamento total por mês")
```

```python
# Gráfico horizontal ordenado 
df_faturamento.sort_values('valor').plot.barh(x='mes', y='valor', title='Faturamento mensal - Ordenado')
```

```python
# Gráfico de Linhas
df_faturamento.plot.line(x='mes', y='valor', title='Faturamento por mês')
```
